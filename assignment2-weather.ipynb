```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/joann/Desktop/PFDA-courseware/assignment/weatherreadings1.csv")
print(df["reportStartDateTime"].head())
print(df["dryBulbTemperature_Celsius"].head())
```

```python
# convert dates from a string into actual dates
df["reportStartDateTime"] = pd.to_datetime(df["reportStartDateTime"])

# choose our relevant columns
df = df[["reportStartDateTime", "dryBulbTemperature_Celsius"]]
```

```python
# removes any missing values that could interfere with the plot
df = df.dropna(subset=["dryBulbTemperature_Celsius"])
```

```python
# make a nice plot
plt.figure(figsize=(10,6))
plt.plot(df["reportStartDateTime"], df["dryBulbTemperature_Celsius"], marker= "o", color = "hotpink", linewidth=2)
plt.title("Temperature Over Time", fontsize=16)
plt.xlabel("Time (MM-DD HH)", fontsize=14)
plt.ylabel("Temperature in Celsius", fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
